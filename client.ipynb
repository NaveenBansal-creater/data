```python
import joblib
sap = joblib.load('SAP_data.pkl')
```

```python
from jina import Client, Document
from jina.types.request import Response


def print_matches(resp: Response):  # the callback function invoked when task is done
    for idx, d in enumerate(resp.docs[0].matches[:5]):  # print top-3 matches
        #print(f'[{idx}]{d.scores["euclidean"].value:2f}: "{d.text}"')
        print (f"{sap[d.text]} | {d.scores['euclidean'].value:2f} | {d.text.upper()}")
        print ("--"*20)
        
def get_results(query):
    c = Client(protocol='grpc', port=12345)  # connect to localhost:12345
    c.post('/search', Document(text=query.lower()), on_done=print_matches)
```

```python
query='Gauge Valve, Male NPT inlet x (3 ea.) 1/2 Female NPT outlets, 6,000psig, Integral seat, 316SS body, bonnet and packing nut, Comply with NACE MR-0175-2009 for Sour Service'        
get_results(query)
```
